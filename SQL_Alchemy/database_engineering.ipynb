```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')

# Import SQLAlchemy `automap` and other dependencies here
import sqlalchemy
from sqlalchemy.orm import Session

from sqlalchemy import create_engine, inspect, func

# Imports the method used for connecting to DBs
from sqlalchemy import create_engine

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float 
```

```python
engine = create_engine("sqlite:///Instructions//Resources/hawaii.sqlite", echo=False)
```

```python
file_one = "Instructions/Resources/clean_hawaii_measurements.csv"

measure_df = pd.read_csv(file_one)
measure_df.head()
```

```python
file_two = "Instructions/Resources/clean_hawaii_stations.csv"
stations_df = pd.read_csv(file_two)
stations_df.head()
```

```python
# Use this to clear out the db
# ----------------------------------
#Base.metadata.drop_all(engine)
#conn.close()
```

```python
session = Session(bind=engine)
```

```python
Base = declarative_base()

class Measurements(Base):
    __tablename__ = 'measurements'
    id = Column(Integer, primary_key=True)
    station = Column(String)
    date = Column(String)
    prcp = Column(Float)
    tobs = Column(Integer)
    
class Stations(Base):
     __tablename__ = 'stations'
     id = Column(Integer, primary_key=True)   
     station = Column(String)
     name = Column(String)
     latitude = Column(Float)
     longitude = Column(Float)
     elevation = Column(Float)
        
conn = engine.connect()
Base.metadata.create_all(engine)

# Use this to clear out the db
# ----------------------------------
# Base.metadata.drop_all(engine)
        
```

```python
for index, row in stations_df.iterrows():
   # print('Index is: {}'.format(index))
    #print('Station is {}; Name is: {}'.format( row['station'], row['name']))
    addRow = Stations(station=row['station'],name=row['name'], latitude=row['latitude'], longitude=row['longitude'], elevation=row['elevation'])
    session.add(addRow)


session.commit()
```

```python
sta_list = session.query(Stations)

for stns in sta_list:
    print(stns.id)
    print(stns.latitude)
```

```python
for index, row in measure_df.iterrows():
   # print('Index is: {}'.format(index))
    #print('Station is {}; Name is: {}'.format( row['station'], row['name']))
    addRow = Measurements(station=row['station'],date=row['date'], prcp=row['prcp'], tobs=row['tobs'])
    session.add(addRow)


session.commit()
```

```python
msr_list = session.query(Measurements)

for msrmnts in msr_list:
    print(f'id: {msrmnts.id}; prcp: {msrmnts.prcp}; tobs: {msrmnts.tobs}')
    
```
